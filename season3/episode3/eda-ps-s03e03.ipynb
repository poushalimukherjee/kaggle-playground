<a id="validation"></a>
## <center><font size = 4><span style="color:#3CB5CD"> <p style="background-color:#EAEFF0;font-family:'Lucida Sans';color:#2B6E7D;font-size:130%;text-align:left;border-radius:5px 5px;padding:5px">2. Categorical Features Validation Following Original IBM Dataset</p></span></font></center>
#### [<font size = 5><span style="color:#202833;font-family:'Times New Roman';font-size:80%;">⬆️</span></font>](#explore-features)

````python
dict_education          = {1:'Below College', 2: 'College', 3: 'Bachelor', 4:'Master', 5:'Doctor'}
dict_env_satisfaction   = {1:'Low', 2:'Medium', 3:'High', 4:'Very High'}
dict_job_involvement    = {1:'Low', 2:'Medium', 3:'High', 4:'Very High'}
dict_job_satisfaction   = {1:'Low', 2:'Medium', 3:'High', 4:'Very High'}
dict_perfm_rating       = {1:'Low', 2:'Goood', 3:'Excellent', 4:'Outstanding'}
dict_reltn_satisfaction = {1:'Low', 2:'Medium', 3:'High', 4:'Very High'}
dict_work_life_bal      = {1:'Bad', 2:'Good', 3:'Better', 4:'Best'}

# Found discrepancy in 'Education' column
mode_edu = int(df.mode(dropna=False)['Education'][0])        
df['Education'] = df['Education'].transform(lambda x: mode_edu if x not in dict_education.keys() else x)
````

<a id="create-features"></a>
## <center><font size = 4><span style="color:#3CB5CD"> <p style="background-color:#EAEFF0;font-family:'Lucida Sans';color:#2B6E7D;font-size:130%;text-align:left;border-radius:5px 5px;padding:5px">3. Create Features or Transform Features as Needed</p></span></font></center>
#### [<font size = 5><span style="color:#202833;font-family:'Times New Roman';font-size:80%;">⬆️</span></font>](#explore-features)

````python
df['Attrition_label'] = df['Attrition'].transform(lambda x: "Left_Company" if x==1 else "Stayed")

df['Education_label'] = df['Education'].transform(lambda x: dict_education[x])
df['EnvironmentSatisfaction_label'] = df['EnvironmentSatisfaction'].transform(lambda x: dict_env_satisfaction[x])
df['JobInvolvement_label'] = df['JobInvolvement'].transform(lambda x: dict_job_involvement[x])
df['JobSatisfaction_label'] = df['JobSatisfaction'].transform(lambda x: dict_job_satisfaction[x])
df['PerformanceRating_label'] = df['PerformanceRating'].transform(lambda x: dict_perfm_rating[x])
df['RelationshipSatisfaction_label'] = df['RelationshipSatisfaction'].transform(lambda x: dict_reltn_satisfaction[x])
df['WorkLifeBalance_label'] = df['WorkLifeBalance'].transform(lambda x: dict_work_life_bal[x])
````

<a id="basic-stats"></a>
## <center><font size = 4><span style="color:#3CB5CD"> <p style="background-color:#EAEFF0;font-family:'Lucida Sans';color:#2B6E7D;font-size:130%;text-align:left;border-radius:5px 5px;padding:5px">4. Some Basic Statistics</p></span></font></center>
#### [<font size = 5><span style="color:#202833;font-family:'Times New Roman';font-size:80%;">⬆️</span></font>](#explore-features)

````python
print(  "Count of instantes of Attrition ( True / False): ", 
             df[  df['Attrition_label'] == "Left_Company" ].shape[0],
      " / ", df[  df['Attrition_label'] == "Stayed"       ].shape[0]
     ,"\n"
     )

columns_cat = ['Gender', 'BusinessTravel', 'Department', 'Education_label', 'EducationField',
               'EnvironmentSatisfaction_label','JobInvolvement_label', 'JobLevel', 'JobSatisfaction_label', 
               'MaritalStatus', 'OverTime', 'PerformanceRating_label',
               'RelationshipSatisfaction_label', 'StockOptionLevel','WorkLifeBalance_label']

for cat_feat in columns_cat:
    print("Feature: ", cat_feat.replace("_label",""), " Attrition ( True / False):", end="")
    df_tmp = df[[cat_feat]]
    df_tmp = df_tmp.drop_duplicates()
    str_len = max([len(str(el)) for el in df_tmp[cat_feat]])
    for el in df_tmp[cat_feat]:
        print("\n------------------> ", " "*(str_len-len(str(el))),el, ": ", 
               df[ (df['Attrition_label'] == "Left_Company") 
                     & (df[cat_feat]  == el ) ].shape[0], "/",               
               df[ (df['Attrition_label'] == "Stayed") 
                     & (df[cat_feat]  == el ) ].shape[0]
              , end="")
    print("\n")
````

<a id="daily-rate"></a>
## <font size = 4><span style="color:#3CB5CD"> <p style="background-color:#FFFFFF;font-family:'Lucida Sans';color:#2B6E7D;font-size:130%;text-align:left;border-radius:5px 5px;padding:5px;margin:0px;margin-top:-20px;">Exploring the _'DailyRate'_ feature with respect to other categorical features</p></span></font>
#### [<font size = 5><p style="font-size:80%;text-align:left;padding:0px;margin:0px;margin-top:0px;">⬆️</p></font>](#explore-numeric)

````python
z = np.abs(stats.zscore(df['DailyRate']))
threshold = 3

outlying_dailyrate_index = np.where(z > threshold)
outlying_dailyrate = [df['DailyRate'][ix] for ix in outlying_dailyrate_index[0]]

df_dummy = df_dummy[ ~df_dummy['DailyRate'].isin(outlying_dailyrate) ]
````

````python
fig1, axes1 = plt.subplots(2,3, figsize=(20,12))
fig1.suptitle("Explore 'DailyRate' Data")

sns.set_theme(style="whitegrid", palette='dark:#5A9_r')
sns.histplot(ax=axes1[0,0], data=df_dummy, x='DailyRate', hue='Attrition_label', palette=pal, kde=True, stat='probability')
axes1[0,0].set_title("Distribution of Daily-Rate WITHOUT Outliers")

sns.set_theme(style="whitegrid", palette='dark:#5A9_r')
sns.scatterplot(ax=axes1[0,1], x=df_dummy['DailyRate'], y=df_dummy['Age'], hue=df_dummy['Attrition_label'], palette=pal)
axes1[0,1].set_title("Scatterplot of Daily-Rate vs Age")

sns.set_theme(style="whitegrid", palette='dark:#5A9_r')
sns.scatterplot(ax=axes1[0,2], x=df_dummy['DailyRate'], y=df_dummy['HourlyRate'], hue=df_dummy['Attrition_label'], palette=pal)
axes1[0,2].set_title("Scatterplot of Daily-Rate vs Hourly-Rate")

sns.set_theme(style="whitegrid", palette='dark:#5A9_r')
sns.scatterplot(ax=axes1[1,0], x=df_dummy['DailyRate'], y=df_dummy['MonthlyRate'], hue=df_dummy['Attrition_label'], palette=pal)
axes1[1,0].set_title("Scatterplot of Daily-Rate vs Monthly-Rate")
````
